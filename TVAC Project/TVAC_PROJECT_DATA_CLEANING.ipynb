```python
import pandas as pd

# Loading CSV file
student_data = pd.read_csv('/content/StudentPerformanceFactors.csv')

# Inspecting the first rows
print(student_data.head())

print(student_data.info())
```

```python
# Cleaning the data
# Checking missing values
print(student_data.isnull().sum())
```

```python
cleaned_data = student_data.dropna()
print(cleaned_data)
print(cleaned_data.isnull().sum())
```

```python
cleaned_data = cleaned_data[cleaned_data['Exam_Score'] != 101]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create boxplots for all numeric columns
numeric_columns = cleaned_data.select_dtypes(include=[np.number]).columns
cleaned_data[numeric_columns].boxplot(figsize=(12, 6), rot=45)
plt.title("Boxplots of Numeric Columns")
plt.ylabel("Values")
plt.show()
```

```python
# Function to detect and drop outliers using IQR
def drop_outliers(data):
    numeric_columns = data.select_dtypes(include=[np.number]).columns  # Select only numeric columns
    for column in numeric_columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Drop rows with outliers in this column
        data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data

# Apply the function to drop outliers
cleaned_no_data = drop_outliers(cleaned_data)

# Print the cleaned data
print("Data after dropping outliers:")
print(cleaned_no_data)
```
